```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_excel("creditcard.xlsx")

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(df['distance_from_home'][:500], df['distance_from_last_transaction'][:500], color='red')
plt.title("Line Plot (distance_from_home vs distance_from_last_transaction)")
plt.xlabel("distance_from_home")
plt.ylabel("distance_from_last_transaction")

plt.subplot(2, 2, 2)
pie_values = df['distance_from_last_transaction'][:5].abs()
plt.pie(pie_values, labels=df['distance_from_home'][:5], autopct="%1.1f%%")
plt.title("Pie Chart (First 5 Values)")

plt.subplot(2, 2, 3)
plt.bar(df['distance_from_home'][:10], df['distance_from_last_transaction'][:10], color='red')
plt.title("Bar Chart (10 Samples)")
plt.xlabel("distance_from_home")
plt.ylabel("distance_from_last_transaction")


plt.subplot(2, 2, 4)
plt.scatter(df['distance_from_home'][:500], df['distance_from_last_transaction'][:500], color='red')
plt.title("Scatter Plot (distance_from_home vs distance_from_last_transaction)")
plt.xlabel("distance_from_home")
plt.ylabel("distance_from_last_transaction")

plt.tight_layout()
plt.savefig("dashboard.png",dpi=300,bbox_inches='tight')
plt.show()
```
